```python
import pandas as pd

def get_dataframe(model, size, config, version="v1"):
    if model == "llama-2":
        if size != 70:
            df = pd.read_json(f"../results/{model}/Llama-2-{size}b-hf/pretrained__meta-llama__Llama-2-{size}b-hf_basquetrivia_{config}.jsonl")
        else:
            df = pd.read_json(f"../results/{model}/Llama-2-{size}b-hf/pretrained__meta-llama__Llama-2-{size}b-hf,parallelize__True_basquetrivia_{config}.jsonl")
    if model == "latxa":
        if size != 70:
            df = pd.read_json(f"../results/{model}/latxa-{size}b-{version}/pretrained__HiTZ__latxa-{size}b-{version}_basquetrivia_{config}.jsonl")
        else:
            df = pd.read_json(f"../results/{model}/latxa-{size}b-{version}/pretrained__HiTZ__latxa-{size}b-{version},parallelize__True_basquetrivia_{config}.jsonl")
    if model == "qwen":
        if size != 72:
            df = pd.read_json(f"../results/{model}/Qwen1.5-{size}B/pretrained__Qwen__Qwen1.5-{size}B_basquetrivia_{config}.jsonl")
        else:
            df = pd.read_json(f"../results/{model}/Qwen1.5-{size}B/pretrained__Qwen__Qwen1.5-{size}B,parallelize__True_basquetrivia_{config}.jsonl")
    if model == "bloom":
        df = pd.read_json(f"../results/{model}/bloom-7b1/pretrained__bigscience__bloom-7b1,dtype__float16_basquetrivia_{config}.jsonl")
    if model == "xglm":
        df = pd.read_json(f"../results/{model}/xglm-7.5B/pretrained__facebook__xglm-7.5B,dtype__float16_basquetrivia_{config}.jsonl")
    if model == "mgpt":
        df = pd.read_json(f"../results/{model}/mGPT-13B/pretrained__ai-forever__mGPT-13B,dtype__float16_basquetrivia_{config}.jsonl")
    if model == "mistral":
        df = pd.read_json(f"../results/{model}/Mistral-7B-v0.1/pretrained__mistralai__Mistral-7B-v0.1_basquetrivia_{config}.jsonl")
    if model == "mixtral":
        df = pd.read_json(f"../results/mistral/Mixtral-8x7B-v0.1/pretrained__mistralai__Mixtral-8x7B-v0.1,parallelize__True_basquetrivia_{config}.jsonl")
    if model == "yi":
        df = pd.read_json(f"../results/{model}/Yi-34B/pretrained__01-ai__Yi-34B,parallelize__True_basquetrivia_{config}.jsonl")
    return df
```

```python
import matplotlib.pyplot as plt

def plot_eus_en_sizes(df_llama2_7b_basquetrivia_eu, df_llama2_13b_basquetrivia_eu, df_llama2_70b_basquetrivia_eu,df_llama2_7b_basquetrivia_en, df_llama2_13b_basquetrivia_en, df_llama2_70b_basquetrivia_en):
    merged = pd.DataFrame()
    merged["D=EU S=7"] = df_llama2_7b_basquetrivia_eu["acc"]
    merged["D=EU S=13"] = df_llama2_13b_basquetrivia_eu["acc"]
    merged["D=EU S=70"] = df_llama2_70b_basquetrivia_eu["acc"]
    merged["D=EN S=7"] = df_llama2_7b_basquetrivia_en["acc"]
    merged["D=EN S=13"] = df_llama2_13b_basquetrivia_en["acc"]
    merged["D=EN S=70"] = df_llama2_70b_basquetrivia_en["acc"]
    
    # plot average accuracy per model
    merged_mean = merged.mean()
    merged_mean.plot(kind="bar", rot=90, title="Average Accuracy per Model")
    display(pd.DataFrame((merged_mean* 100).round(2)))

    merged["group"] = df_llama2_7b_basquetrivia_eu["doc"].apply(lambda x: x["group"] if x["group"] != "Nazioarteko gaiak" else "Gai orokorrak")
    merged["difficulty"] = df_llama2_7b_basquetrivia_eu["doc"].apply(lambda x: x["difficulty"])
    merged["category"] = df_llama2_7b_basquetrivia_eu["doc"].apply(lambda x: x["category"])

    merged_grouped = merged.groupby("group").mean()
    merged_grouped_difficulty = merged.groupby("difficulty").mean()
    merged_grouped_category = merged.groupby("category").mean()
    # remove difficulty from merged_grouped and merged_grouped_category
    merged_grouped.drop("difficulty", axis=1, inplace=True)
    merged_grouped_category.drop("difficulty", axis=1, inplace=True)

    fig, axs = plt.subplots(1, 3, figsize=(20, 6))
    merged_grouped.plot(kind="bar", ax=axs[0], rot=0, title="Accuracy by Group")
    merged_grouped_difficulty.plot(kind="bar", ax=axs[2], rot=0, title="Accuracy by Difficulty")
    merged_grouped_category.plot(kind="bar", ax=axs[1], rot=90, title="Accuracy by Category")
    
    # multiply by 100 and round to 1 decimal place
    display((merged_grouped.T * 100).round(2))
    display((merged_grouped_difficulty.T * 100).round(2))
    display((merged_grouped_category.T * 100).round(2))
    
    # add plot that combines group and category
    merged_group_category = merged.groupby(["category", "group"]).mean()
    merged_group_category.drop("difficulty", axis=1, inplace=True)
    # merged_group_category.plot(kind="bar", rot=90, title="Accuracy by Category and Group")
    # make this plot bigger
    plt.gcf().set_size_inches(20, 6)
    display((merged_group_category.T * 100).round(2))
```

```python
plot_eus_en_sizes(df_llama2_7b_basquetrivia_eu, df_llama2_13b_basquetrivia_eu, df_llama2_70b_basquetrivia_eu,df_llama2_7b_basquetrivia_en, df_llama2_13b_basquetrivia_en, df_llama2_70b_basquetrivia_en)
```
